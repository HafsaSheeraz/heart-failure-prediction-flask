*Hyperparameter tuning*


```python
# Splitting dataset into X and Y
# X = all columns except 'DEATH_EVENT'
X = dataset.drop('DEATH_EVENT', axis=1)

# y = target column
Y = dataset['DEATH_EVENT']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split (if not done already)
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)

# Initialize base model
model = LogisticRegression(solver='liblinear')  # Use 'liblinear' to support l1 penalty

# Hyperparameter tuning with GridSearchCV
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
}

grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, Y_train)

# Best model from grid search
best_model = grid_search.best_estimator_

# Evaluation
print("✅ Best Parameters:", grid_search.best_params_)
print("✅ Best Cross-Validation Accuracy:", grid_search.best_score_)

train_preds = best_model.predict(X_train)
train_accuracy = accuracy_score(Y_train, train_preds)
print("✅ Training Accuracy:", train_accuracy)

test_preds = best_model.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_preds)
print("✅ Test Accuracy:", test_accuracy)

# Predict for a new input sample
input_sample = [[63,	1,	61,	1,	40,	0,	221000,	1.1,	140,	0,	0,	86]]
input_sample_scaled = scaler.transform(input_sample)
prediction = best_model.predict(input_sample_scaled)
result = "Died" if prediction[0] == 1 else "Not Died"
print("✅ Prediction for input sample:", result)

```
